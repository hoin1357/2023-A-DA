```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[10,20,30,40],label='ascending')
plt.plot([1,2,3,4],[40,30,20,10],label='descending')

plt.title('legend')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[10,20,30,40],color='skyblue',label='ascending')
plt.plot([1,2,3,4],[40,30,20,10],"pink",label='descending')

plt.title('Color')
plt.title('legend')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[10,20,30,40],color='skyblue',linestyle='--',label='dashed')
plt.plot([1,2,3,4],[40,30,20,10], 'b', ls=":",label='dotted')

plt.title('linestyle')
plt.title('legend')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[10,20,30,40],color='skyblue',linestyle='--',label='diamond',marker='d')
plt.plot([1,2,3,4],[40,30,20,10], 'b', ls=":",marker='^',label='triangle up')

plt.title('marker')
plt.title('legend')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[10,20,30,40],color='r',linestyle='--',marker='d',label='thick line',lw=3,)
plt.plot([1,2,3,4],[40,30,20,10], 'b',marker='^', ls=":",label='thick line',lw=0.5)

plt.title('marker')
plt.title('legend')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
import matplotlib as mpl
from matplotlib import font_manager

font_list=sorted([f.name for f in font_manager.fontManager.ttflist ])
"Malgun Gothic" in font_list
```

```python
import matplotlib.pyplot as plt

plt.rc("font",family="Malgun Gothic")
plt.rcParams['axes.unicode_minus'] = False

plt.plot([-2,-1,0,1,2,3],[20,40,20,40,20,40],label='선 1')
plt.plot([-2,-1,0,1,2,3],[20,40,20,40,20,40],label='선2')

plt.title('marker')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
import numpy as np

x=np.arange(-10,10,0.1)
plt.plot(x,x**2)

plt.xticks(np.arange(-10,11,2))
plt.yticks(np.arange(0,101,20))

plt.rc("font",family="Malgun Gothic")
plt.title('하기싫어')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,2*np.pi,200)
y=np.sin(x)
plt.plot(x,y)

plt.grid(True)
plt.rc('font',family="Malgun Gothic")
plt.rcParams["axes unicide_minus"]=False

plt.title('하기싫어')
plt.show()
```
