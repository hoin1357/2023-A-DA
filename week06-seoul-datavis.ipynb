```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)

header=next(data)

mean=[]
low=[]
high=[]

for row in data:
    if (row[2] !='')and(row[3] !='')and(row[4] !=''):
        if (row[0].split('-')[1]=='12') and (row[0].split('-')[2]=='25'):
            mean.append(float(row[2]))
            low.append(float(row[3]))
            high.append(float(row[4]))

f.close()

plt.figure(figsize=(10,4))
plt.plot(mean,'k')
plt.plot(low,'b')
plt.plot(high,'r')
plt.xlabel('Day(December 25)')
plt.ylabel('The highest temperature')
plt.rc('font',family='Malgun Gothic')
plt.title('크리스마스의 기온 변화 그래프')
plt.rcParams['axes.unicode_minus']=False
plt.legend()
plt.grid()
plt.show()




```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)

header=next(data)

temp_day=[]
temp_mean=[]
year=[]

for row in data:
    if int(row[0].split('-')[0])>1907:
        if row[2] !='':
            temp_day.append(float(row[2]))
            if(row[0].split('-')[1]=='12')and(row[0].split('-')[2]=='31'):
                year.append(int(row[0].split('-')[0]))
                temp_mean.append(sum(temp_day)/len(temp_day))
                temp_day=[]
f.close()

plt.figure(figsize=(10,4))
plt.rc('font',family='Malgun Gothic')
plt.title('평균 기온 변화 그래프')
plt.plot(year,temp_mean,color='g',marker='d')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.grid()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)

header=next(data)
jan=[]
aug=[]

for row in data:
    month=row[0].split('-')[1]
    if row[4]!='':
      if month=='01':
         jan.append(float(row[4]))
      elif month=='08':
         aug.append(float(row[4]))

plt.boxplot([jan,aug])
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)

header=next(data)

day=[]
for i in range(31):
    day.append([])

for row in data:
    if row[4]!='':
        if row[0].split('-')[1]=='01':
            day[int(row[0].split('-')[2])-1].append(float(row[4]))

plt.style.use('ggplot')
plt.figure(figsize=(10,5))
plt.boxplot(day,showfliers=False)
plt.show()
```
