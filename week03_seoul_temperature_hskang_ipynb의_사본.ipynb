```python
import csv

max_temp=-999
max_date=''

f=open('seoul.csv','r',encoding='cp949')
data=csv.reader(f)
header=next(data)

print(header)
for row in data:
    if row[4]=='':
        row[4]=-999

    row[4]=float(row[4])

    if max_temp<row[4]:
        max_temp=row[4]
        max_date=row[0].strip()

f.close()

print(f'관측  이래 서울의 최고기온을 가졌던 날은{max_date}로{max_temp}였습니다.')
```
