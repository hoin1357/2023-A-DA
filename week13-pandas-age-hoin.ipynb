## 올림픽 메달 집계 분석

```python
import pandas as pd 

df=pd.read_html("https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84")
print(df)
```

```python
type(df)
```

```python
len(df)
```

```python
df[0]
```

```python
df2=df[0].set_index('국가 (IOC 코드)')
df2
```

```python
summer=df2.iloc[:, :5]
summer
```

```python
summer.iloc[3:10,2:4]
```

```python
summer=df2.iloc[:, :5]
summer.columns=['경수기','금','은','동','힙계']
summer
```

```python
summer=df2.iloc[:, :5]
summer.columns=['경수기','금','은','동','힙계']
summer=summer.sort_values('금',ascending=False)
summer
```

```python
summer=df2.iloc[:, :5]
summer.columns=['경수기','금','은','동','힙계']
summer=summer.sort_values('금',ascending=False)
summer.to_excel('하계올림픽메달.xlsx')
```
